```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

# Resizing Graph
plt.figure(figsize = (5,3), dpi = 100)
# plt.plot(x,y,label='2x', color = 'r', linewidth = 2, marker = '.', linestyle = '--', markersize = 10, markeredgecolor = 'blue')


# using shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'r^--',label='2x')

# Second line
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, label = 'x^2')
plt.plot(x2[4:],x2[4:]**2, 'g--')

plt.title('My first graph', fontdict ={'fontname': 'Comic Sans MS','fontsize' : 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks(x)
plt.yticks(y)

plt.legend()
plt.savefig('mygraph.png', dpi = 600)
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,2,3]

bars = plt.bar(labels, values)

patterns = ['/', 'O','*']
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize = (6,4))
plt.show()
```

## FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

### HISTOGRAM

```python
bins = [40,50,60,70,80,90,100] 
plt.hist(fifa.Overall, bins = bins)
plt.ylabel('Number of players')
plt.xlabel('Skill level')
plt.title('Distribution of PLayer skills in FIFA ')
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
lables= ['Left', 'Right'] 
plt.pie([left,right], labels = lables, autopct = '%.2f%%')
plt.title('Foot Preference of FIFA players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]
plt.figure(figsize=(10,10))
weights= [light, light_medium, medium, medium_heavy, heavy] 
labels = ['Under 125', '125-150','150-175','175-200', 'Over 200']
explode = (.4,.1,0,0,.4)
plt.pie(weights, labels = labels, autopct = '%.2f%%', explode = explode)
plt.title('Weight Distribution of FIFA players (in lbs)')
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club =='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club =='Real Madrid']['Overall']
revs = fifa.loc[fifa.Club =='New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid','New England Revolution']
    
plt.figure(figsize=(10,10))
boxes = plt.boxplot([barcelona, madrid, revs], labels = labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
#     Setting edge color
    box.set(color='#24fc03', linewidth = 2)
    
#  Setting Fill color
    box.set(facecolor='#fcf003')

plt.title('Professional Soccer team comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
