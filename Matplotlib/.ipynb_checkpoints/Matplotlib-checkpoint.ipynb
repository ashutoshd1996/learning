```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

# Resizing Graph
plt.figure(figsize = (5,3), dpi = 100)
# plt.plot(x,y,label='2x', color = 'r', linewidth = 2, marker = '.', linestyle = '--', markersize = 10, markeredgecolor = 'blue')


# using shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'r^--',label='2x')

# Second line
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, label = 'x^2')
plt.plot(x2[4:],x2[4:]**2, 'g--')

plt.title('My first graph', fontdict ={'fontname': 'Comic Sans MS','fontsize' : 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks(x)
plt.yticks(y)

plt.legend()
plt.savefig('mygraph.png', dpi = 600)
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,2,3]

bars = plt.bar(labels, values)

patterns = ['/', 'O','*']
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize = (6,4))
plt.show()
```
